```python
import numpy as np

def track_players(results_video):
    tracked_players = {}
    player_id_counter = 0

    for frame_index, results in enumerate(results_video):
        boxes = results.boxes.xyxy.cpu().numpy()

        for box in boxes:
            x1, y1, x2, y2 = box
            centroid = ((x1 + x2) / 2, (y1 + y2) / 2)
            match_found = False

            for player_id, player_data in tracked_players.items():
                prev_centroid = player_data['positions'][-1]
                distance = np.linalg.norm(np.array(centroid) - np.array(prev_centroid))

                if distance < 50:
                    tracked_players[player_id]['positions'].append(centroid)
                    tracked_players[player_id]['last_seen_frame'] = frame_index
                    match_found = True
                    break

            if not match_found:
                tracked_players[player_id_counter] = {
                    'positions': [centroid],
                    'last_seen_frame': frame_index
                }
                player_id_counter += 1

    return tracked_players
```

```python
import numpy as np
from scipy.spatial.distance import euclidean
from scipy.optimize import linear_sum_assignment  # ✅ NEW

def calculate_trajectory_similarity(positions_A, positions_B):
    """
    Calculate similarity between two trajectories.
    Assumes both have same number of positions (aligned by frame).
    """
    if len(positions_A) != len(positions_B):
        min_len = min(len(positions_A), len(positions_B))
        positions_A = positions_A[:min_len]
        positions_B = positions_B[:min_len]

    distances = [euclidean(p1, p2) for p1, p2 in zip(positions_A, positions_B)]
    return np.mean(distances)

def match_players(tracked_players_A, tracked_players_B):
    """
    Match players using Hungarian algorithm based on average trajectory distance.
    """
    ids_A = list(tracked_players_A.keys())
    ids_B = list(tracked_players_B.keys())

    cost_matrix = np.zeros((len(ids_A), len(ids_B)))

    for i, id_a in enumerate(ids_A):
        for j, id_b in enumerate(ids_B):
            pos_a = tracked_players_A[id_a]['positions']
            pos_b = tracked_players_B[id_b]['positions']
            cost = calculate_trajectory_similarity(pos_a, pos_b)
            cost_matrix[i, j] = cost

    row_ind, col_ind = linear_sum_assignment(cost_matrix)

    matched_pairs = {}
    for i, j in zip(row_ind, col_ind):
        if cost_matrix[i, j] < 100:
            matched_pairs[ids_A[i]] = ids_B[j]

    return matched_pairs
```

```python
def calculate_distances(positions):
    """
    Calculate distance between consecutive positions (per frame).
    Returns a list of distances.
    """
    distances = []
    for i in range(1, len(positions)):
        d = np.linalg.norm(np.array(positions[i]) - np.array(positions[i - 1]))
        distances.append(d)
    return distances

def calculate_speeds(distances, frame_rate=30):
    """
    Estimate speed (pixels per second) from distances and frame rate.
    """
    time_interval = 1 / frame_rate  # seconds per frame
    speeds = [d / time_interval for d in distances]
    return speeds

def analyze_trajectories(matched_players, tracked_A, tracked_B, frame_rate=30):
    """
    Analyze matched trajectories between two videos.
    Returns per-player stats (distance covered, average speed).
    """
    analysis = {}

    for id_A, id_B in matched_players.items():
        positions_A = tracked_A[id_A]['positions']
        positions_B = tracked_B[id_B]['positions']


        distances = calculate_distances(positions_A)
        total_distance = sum(distances)
        speeds = calculate_speeds(distances, frame_rate)
        avg_speed = np.mean(speeds)

        analysis[id_A] = {
            'matched_id_B': id_B,
            'total_distance_px': total_distance,
            'avg_speed_px_per_sec': avg_speed,
            'frame_count': len(positions_A),
        }

    return analysis
```

```python
import matplotlib.pyplot as plt

def plot_trajectories(tracked_players, player_ids=None, field_size=(640, 360)):
    """
    Plot player trajectories on a field.
    """
    plt.figure(figsize=(10, 6))
    plt.title("Player Trajectories on Field")
    plt.xlim(0, field_size[0])
    plt.ylim(0, field_size[1])
    plt.gca().invert_yaxis()  # Match video coordinate system

    if player_ids is None:
        player_ids = tracked_players.keys()

    for player_id in player_ids:
        positions = tracked_players[player_id]['positions']
        xs, ys = zip(*positions)
        plt.plot(xs, ys, marker='o', label=f'Player {player_id}')

    plt.xlabel("X Position (pixels)")
    plt.ylabel("Y Position (pixels)")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def calculate_distances(positions):
    distances = []
    for i in range(1, len(positions)):
        dist = np.linalg.norm(np.array(positions[i]) - np.array(positions[i - 1]))
        distances.append(dist)
    return distances

def calculate_speeds(distances, fps=30):  # Assume 30 FPS if unknown
    return [d * fps for d in distances]

def plot_speed_curve(speeds, player_id):
    if len(speeds) < 2:
        print(f"⚠️ Not enough speed data for Player {player_id}")
        return

    plt.figure(figsize=(8, 4))
    plt.plot(speeds, label=f"Player {player_id}", color='blue')
    plt.xlabel("Time (frames)")
    plt.ylabel("Speed (px/s)")
    plt.title(f"🚀 Speed Curve - Player {player_id}")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def plot_heatmap(tracked_players, player_id):
    player = tracked_players.get(player_id)
    if not player or len(player['positions']) < 5:
        print(f"⚠️ Not enough data to plot heatmap for Player {player_id}")
        return

    positions = np.array(player['positions'])
    x, y = positions[:, 0], positions[:, 1]

    heatmap, xedges, yedges = np.histogram2d(x, y, bins=50)
    heatmap = heatmap.T  # flip for orientation

    plt.figure(figsize=(8, 6))
    sns.heatmap(heatmap, cmap='magma', cbar=True)
    plt.title(f"📍 Heatmap for Player {player_id}")
    plt.xlabel("X Position")
    plt.ylabel("Y Position")
    plt.tight_layout()
    plt.show()

```

```python
import cv2

def draw_trajectories_on_video(video_path, tracked_players, output_path="output.mp4"):
    cap = cv2.VideoCapture(video_path)
    width  = int(cap.get(3))
    height = int(cap.get(4))
    fps    = cap.get(cv2.CAP_PROP_FPS)

    out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

    frame_index = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        for player_id, data in tracked_players.items():
            if frame_index < len(data['positions']):
                x, y = map(int, data['positions'][frame_index])
                cv2.circle(frame, (x, y), 5, (0, 255, 0), -1)
                cv2.putText(frame, f'{player_id}', (x+5, y), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,255), 1)

        out.write(frame)
        frame_index += 1

    cap.release()
    out.release()
    print(f"Saved visualized video to {output_path}")


```

```python
# Step 1: Track players in both videos
tracked_broadcast = track_players(results_broadcast)
tracked_tacticam = track_players(results_tacticam)

# Step 2: Match players
matched = match_players(tracked_broadcast, tracked_tacticam)

# Step 3: Analyze matched trajectories
analysis = analyze_trajectories(matched, tracked_broadcast, tracked_tacticam)

# Step 4: Visualize stats
for player_id in matched:
    speeds = calculate_speeds(calculate_distances(tracked_broadcast[player_id]['positions']))
    plot_speed_curve(speeds, player_id)
    plot_heatmap(tracked_broadcast, player_id)

# Step 5: Optional – Draw player positions on the video
draw_trajectories_on_video("broadcast.mp4", tracked_broadcast, "output_broadcast.mp4")
draw_trajectories_on_video("tacticam.mp4", tracked_tacticam, "output_tacticam.mp4")

# Optional – Save stats
import pandas as pd
df = pd.DataFrame.from_dict(analysis, orient='index')
df.to_csv("player_stats.csv")
```
